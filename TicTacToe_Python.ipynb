Factorial in Python

```python
def factorial(num):
    if not isinstance(num, int) or num < 0:
        raise ValueError("Factorial is defined only for non-negative integers")
    elif num == 0:
        return 1
    else:
        fact = 1
        for i in range(1, num + 1):
            fact = fact * i
        return fact

num_input = input("Input a number to get its factorial? ")
try:
    num = int(num_input)
    result = factorial(num)
    print(f"The factorial of {num} is {result}")
except ValueError as e:
    print(e)
```

Multiplayer Tic Tac Toe Game Written In Python





```python
'''
Tic Tac Toe Game

This would be the flow of the algorithm/ pseudocode for the Tic Tac Toe game.

- Collect and display player names
- Display the game board
- Ask first and second player to make their move
- Check if the move is valid
- Game logic function
- Create a list/ array of 9 variables
- Create validation logic: Create possible winning chances and else player loses
- Create Variable to store each player's moves

'''


def playGame(player1, player2):

    boardInput = ["a", "b", "c", "d", "e", "f", "g", "h", "i"]
    #Starts the program
    while True:
        turn = 0 #initializes the player turn
        while turn < 9:

            if turn % 2 == 0:
              current_player = player1
              marker = 'X'
            else:
              current_player = player2
              marker = 'O'

            print(f'''
            {boardInput[0]} - {boardInput[1]} - {boardInput[2]}\n
            {boardInput[3]} - {boardInput[4]} - {boardInput[5]}\n
            {boardInput[6]} - {boardInput[7]} - {boardInput[8]}\n
            ''')

            playerInput = input(f'{current_player} ({marker}), enter your position (a-i): ')
            playerInput = playerInput.lower().strip()
            turn += 1

            #checks if
            if playerInput == 'a':
              boardInput[0] = marker
            elif playerInput == 'b':
              boardInput[1] = marker
            elif playerInput == 'c':
              boardInput[2] = marker
            elif playerInput == 'd':
              boardInput[3] = marker
            elif playerInput == 'e':
              boardInput[4] = marker
            elif playerInput == 'f':
              boardInput[5] = marker
            elif playerInput == 'g':
              boardInput[6] = marker
            elif playerInput == 'h':
              boardInput[7] = marker
            elif playerInput == 'i':
              boardInput[8] = marker
            else:
              print("Invalid input. Please enter a valid position.")
              continue


            if ((boardInput[0] == boardInput[1] == boardInput[2]) or
                (boardInput[3] == boardInput[4] == boardInput[5]) or
                (boardInput[6] == boardInput[7] == boardInput[8]) or

                (boardInput[0] == boardInput[3] == boardInput[6]) or
                (boardInput[1] == boardInput[4] == boardInput[7]) or
                (boardInput[2] == boardInput[5] == boardInput[8]) or

                (boardInput[0] == boardInput[4] == boardInput[8]) or
                (boardInput[2] == boardInput[4] == boardInput[6])):

                print(f'{current_player}' + " Wins!")
                break

            elif turn == 9:
              print("It's a Draw!")
              break
        break
    else:
      print("Game Over")

print("Welcome to Tic Tac Toe Game!")
player1 = input("Player One please enter your name: ")
player2 = input("Player Two please enter your name:")
print("\n")

print(f"{player1} will be 'X' and {player2} will be 'O'\n")

input("Press enter to start the game")
print("Game started")


playGame(player1, player2)
```
